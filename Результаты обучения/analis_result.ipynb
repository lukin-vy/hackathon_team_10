```python
#Подключим библиотеки
import pandas as pd
import plotly
import plotly.express as px
import plotly.graph_objs as go
```

Проанализируем модель vgg

```python
#Прочитаем данные
vgg_df = pd.read_csv('vgg_lern_result.csv')
#Выведем признаки
print(vgg_df.info())
```

```python
px.line(
    data_frame=vgg_df,
    x='epochs',
    y=['loss_train'],
    height=500,
    width=1000,
    title='loss_train'
).show()
```

```python
px.line(
    data_frame=vgg_df,
    x='epochs',
    y=['loss_train'],
    height=500,
    width=1000,
    title='loss_train'
).show()
```

```python
px.line(
    data_frame=vgg_df,
    x='epochs',
    y=['loss_valid'],
    height=500,
    width=1000,
    title='loss_valid'
).show()
```

```python
px.line(
    data_frame=vgg_df,
    x='epochs',
    y=['acc_valid'],
    height=500,
    width=1000,
    title='acc_valid'
).show()
```

Проанализируем модель google

```python
#Прочитаем данные
google_df = pd.read_csv('google_lern_result.csv')
#Выведем признаки
print(google_df.info())
```

```python
px.line(
    data_frame=google_df,
    x='epochs',
    y=['loss_train'],
    height=500,
    width=1000,
    title='loss_train'
).show()
```

```python
px.line(
    data_frame=google_df,
    x='epochs',
    y=['loss_train'],
    height=500,
    width=1000,
    title='loss_train'
).show()
```

```python
px.line(
    data_frame=google_df,
    x='epochs',
    y=['loss_valid'],
    height=500,
    width=1000,
    title='loss_valid'
).show()
```

```python
px.line(
    data_frame=google_df,
    x='epochs',
    y=['acc_valid'],
    height=500,
    width=1000,
    title='acc_valid'
).show()
```

Сравненеи моделей google и vgg

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=vgg_df['epochs'], y=vgg_df['loss_train'], name='vgg'))
fig.add_trace(go.Scatter(x=google_df['epochs'], y=google_df['loss_train'], name='google'))
fig.update_layout(title="loss_train")
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=vgg_df['epochs'], y=vgg_df['loss_valid'], name='vgg'))
fig.add_trace(go.Scatter(x=google_df['epochs'], y=google_df['loss_valid'], name='google'))
fig.update_layout(title="loss_valid")
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=vgg_df['epochs'], y=vgg_df['acc_train'], name='vgg'))
fig.add_trace(go.Scatter(x=google_df['epochs'], y=google_df['acc_train'], name='google'))
fig.update_layout(title="acc_train")
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=vgg_df['epochs'], y=vgg_df['acc_valid'], name='vgg'))
fig.add_trace(go.Scatter(x=google_df['epochs'], y=google_df['acc_valid'], name='google'))
fig.update_layout(title="acc_valid")
fig.show()
```
